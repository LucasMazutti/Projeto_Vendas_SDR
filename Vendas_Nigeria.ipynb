```python
import pandas as pd 
```

```python
df = pd.read_csv('nigeria_messy_sales_dataset.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Preencher valores nulos nas colunas de texto
df['Customer Name'] = df['Customer Name'].fillna('No name')
df['Sales Channel'] = df['Sales Channel'].fillna('No channel')
df['Order ID'] = df['Order ID'].fillna('No order ID')
```

```python
df.head()
```

```python
print('Valores nulos preenchidos com sucesso nas colunas text')
```

```python
# Converter coluna 'date' para formato datetime
df['Sale Date'] = pd.to_datetime(df['Sale Date'], errors='coerce')
```

```python
df.info()
```

```python
df.dtypes
```

```python
# Colocar valores de 'state' e 'product' em maiúsculo
df['State'] = df['State'].str.upper()
df['Product'] = df['Product'].str.upper()
```

```python
df.head()
```

```python
df.to_csv("dados_limpos.csv", index=False, encoding="utf-8")
```
